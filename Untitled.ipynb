# Explore ICUs measures available
A bag of measures collected from ICUs beds have been integrated with several other tables in a Drill datasource.<br>
The aim of this notebook is to explore this data and see how it can be used to computed additional features which can be used in mortality prediction.

```python
def df_from_query(conn, query):
    """
    Query an ODBC datasource and set results in a pandas Dataframe
    
    Parameters
    ----------
    pyodbc.Connection conn : Datasource ODBC connection
    string           query : query to execute on datasource
    """
    c = conn.cursor()
    c.execute(query)
    cols = [column[0] for column in c.description]
    data = []
    for row in c.fetchall():
        data.append(tuple(row))
    df = pd.DataFrame(data, columns=cols)
    return df
```

```python
#
# Drillbit ODBC connection
#
conn = pyodbc.connect("DSN=drill_eds", autocommit=True)
```

```python
#
# Most frequent types of measures between H0 and H24 in ICUs
#
q1 = "select count(1) as co,s.id_measure_type, r.label \
      from icu_sensors s, nda_mouv_ufr_tr n, icu_ufr u, ref_measure r \
      where cast(s.id_nda as VARCHAR)=n.id_nda and \
      s.dt_insert >= n.dt_deb_mouv_ufr and \
      s.dt_insert <= n.dt_deb_mouv_ufr + interval '1' DAY(2) and \
      u.ids_ufr=n.ids_ufr and \
      s.id_measure_type = cast(r.code as INT) \
      group by s.id_measure_type, r.label \
      order by co desc, s.id_measure_type, r.label"
```

```python
df = df_from_query(conn, q1)
```

```python
df['label'] = df['label'].str.decode('iso-8859-1').str.encode('utf8')
df.head(20)
```

```python
df_plot = df[[u'label', u'co']].head(20)
df_plot.columns = ['measure', 'count']
df_plot.plot(kind='barh', x='measure', y='count', figsize=(15, 10), color="#a54456", alpha=0.7)
```

```python
#
# How many measure type with counter > 12000 ?
#
df[df[u'co'] > 12000].count()
```

```python
#
# Keep only measure type of interest
#
df = df[df[u'co'] > 12000]
df[[u'id_measure_type']].T
```

```python
import re
regex = re.compile(r"[\s]+")
list_measures = regex.sub(",",np.array_str(df[u'id_measure_type'].values))
list_measures = list_measures.replace("[","(").replace("]",")")
list_measures
```

```python
#
# How many ICUs stays (with H0-24 measures) ?
#
q2 = "select count(distinct s.id_nda) \
    from icu_sensors s, nda_mouv_ufr_tr n, icu_ufr u \
    where cast(s.id_nda as VARCHAR)=n.id_nda \
    and s.dt_insert >= n.dt_deb_mouv_ufr \
    and s.dt_insert <= n.dt_deb_mouv_ufr + interval '1' DAY(2) \
    and u.ids_ufr=n.ids_ufr"

q2 += " and s.id_measure_type in "+list_measures

df_from_query(conn, q2)
```

```python
#
# Number of measures by stay and measure type (only 10102, 10120, 11, 12, 14, 15)
#
q3 = "select s.id_nda, s.id_measure_type, count(1) as co \
      from icu_sensors s, nda_mouv_ufr_tr n, icu_ufr u \
      where cast(s.id_nda as VARCHAR)=n.id_nda and \
      s.dt_insert >= n.dt_deb_mouv_ufr and \
      s.dt_insert <= n.dt_deb_mouv_ufr + interval '1' DAY(2) and \
      u.ids_ufr=n.ids_ufr"
# q3 += " and s.id_measure_type in "+list_measures
q3 += " and s.id_measure_type in (10102, 10120, 11, 12, 14, 15)"
q3 += " group by s.id_nda, s.id_measure_type \
      order by s.id_nda, s.id_measure_type"

df_feat = df_from_query(conn, q3)
df_feat.head()
```
